# Inspect MSAS_TP_k01_01.pkl

```python
from pathlib import Path
from batteryml.data.battery_data import BatteryData

pkl_path = Path('data\processed\MSAS\MSAS_S1_TP_z01_01.pkl')
if not pkl_path.exists():
    raise FileNotFoundError(f'File not found: {pkl_path.resolve()}')
```

```python
cell = BatteryData.load(pkl_path)
cell.print_description()
print(f'cycle count: {len(cell.cycle_data)}')
first_cycle = cell.cycle_data[0]
print('first cycle keys:', list(first_cycle.to_dict().keys()))
print('first cycle cycle_number:', first_cycle.cycle_number)
```

```python
def _length(value):
    return len(value) if value is not None else 0

print('voltage points:', _length(first_cycle.voltage_in_V))
print('current points:', _length(first_cycle.current_in_A))
print('charge capacity points:', _length(first_cycle.charge_capacity_in_Ah))
print('discharge capacity points:', _length(first_cycle.discharge_capacity_in_Ah))
print('time points:', _length(first_cycle.time_in_s))
print('temperature points:', _length(first_cycle.temperature_in_C))
```
